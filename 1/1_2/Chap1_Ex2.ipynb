```python
# 3.
prices = df['price'].loc[df['price'] != '?']  #df['price'].str.isnumeric() == True
pmedian = prices.astype(float).median()
df['price'] = df['price'].replace('?', pmedian).astype(float)
```

```python
# 4. thực hiện tương tự cho cột horsepower và nomolized-losses
```

```python
# 5. mean, median và mode
mean = df['height'].mean()
median = df['height'].median()
mode = df['height'].mode()
print(round(mean, 2), median, mode)
```

```python
mean = df['price'].mean()
median = df['price'].median()
mode = df['price'].mode()
print(round(mean, 2), median, mode)
```
