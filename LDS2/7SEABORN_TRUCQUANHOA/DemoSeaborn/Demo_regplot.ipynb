```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/vivekg0510/04.-Data-Visualization-With-Seaborn/master/college_datav3.csv',index_col=0)
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Regions'].value_counts()
```

```python
plt.figure(figsize=(8,6))
sb.regplot(data=df,y='Tuition',x='SAT_AVG_ALL',marker='^',color='g')
plt.show()
```

```python
# display the residual plot
plt.figure(figsize=(8,6))
sb.residplot(data=df,y='Tuition',x='SAT_AVG_ALL',color='g')
plt.show()
```

```python
#polynominal regression
plt.figure(figsize=(8,6))
sb.regplot(data=df,y='SAT_AVG_ALL',x='PCTPELL',order=2)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sb.residplot(data=df,y='SAT_AVG_ALL',x='PCTPELL',order=2)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sb.regplot(data=df,y='Tuition',x='UG',x_bins=8)
plt.show()
```
